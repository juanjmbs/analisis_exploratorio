##EVALUACION DE MODELO 2

se requiere evaluar por lo menos 3 modelos distintos de regresión , optimizar sus hiperparámetros  y seleccionar aquel con menor  error cuadrático medio en una en una validación cruzada en 10 partes.

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
import numpy as np

# Definimos los modelos que vamos a utilizar
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Random Forest Regression': RandomForestRegressor()
}

# Ajustamos los hiperparámetros para Ridge y Lasso
param_grid = {
    'Ridge Regression': {'alpha': [0.1, 1, 10]},
    'Lasso Regression': {'alpha': [0.1, 1, 10]}
}

# Realizamos una validación cruzada en 10 partes para cada modelo
for name, model in models.items():
    if name in param_grid:
        grid_search = GridSearchCV(model, param_grid[name], cv=10, scoring='neg_mean_squared_error')
        grid_search.fit(X_train, y_train)
        best_model = grid_search.best_estimator_
    else:
        best_model = model

    # Calculamos el MSE promedio en validación cruzada
    scores = cross_val_score(best_model, X_train, y_train, cv=10, scoring='neg_mean_squared_error')
    rmse_scores = np.sqrt(-scores)
    print(f'Modelo: {name}')
    print(f'MSE Promedio: {np.mean(rmse_scores)}')

# Seleccionamos el modelo con el menor MSE promedio
best_model_name = min(models, key=lambda x: np.mean(np.sqrt(-cross_val_score(models[x], X_train, y_train, cv=10, scoring='neg_mean_squared_error'))))
best_model = models[best_model_name]
print(f'Mejor modelo: {best_model_name}')

# Entrenamos el mejor modelo con todos los datos de entrenamiento
best_model.fit(X_train, y_train)

# Calculamos el MSE en el conjunto de prueba
y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'MSE en conjunto de prueba: {mse}')
print(f'RMSE en conjunto de prueba: {rmse}')

# Obtenemos la importancia de las características (solo para Random Forest)
if isinstance(best_model, RandomForestRegressor):
    feature_importances = best_model.feature_importances_
    print('Importancia de las características:')
    for i, importance in enumerate(feature_importances):
        print(f'Feature {i}: {importance}')
```

## N-GRAMAS

Los n-gramas se utilizan en sistemas de corrección ortográfica para sugerir palabras correctas a partir de palabras mal escritas. Al comparar la palabra mal escrita con los n-gramas del vocabulario, se pueden identificar las palabras más similares y probables como sugerencias.

```python
import pandas as pd
from nltk.util import ngrams
import re


# Preprocesar texto
def preprocesar_texto(texto):
  texto_minusculas = texto.lower()
  texto_sin_puntuacion = re.sub(r'[^\w\s]', '', texto_minusculas)
  tokens = texto_sin_puntuacion.split()
  return tokens

mt["title_procesado"] = mt["title"].apply(preprocesar_texto)
mt["description_procesado"] = mt["description"].apply(preprocesar_texto)

def generar_ngramas(texto_procesado, n):
  ngramas_lista = []
  for tokens in texto_procesado:
    ngramas_lista.extend(ngrams(tokens, n))
  return ngramas_lista

bigramas_title = generar_ngramas(mt["title_procesado"], 2)
bigramas_description = generar_ngramas(mt["description_procesado"], 2)

# Almacenar bigramas en DataFrames (without specifying columns)
bigramas_title_df = pd.DataFrame(bigramas_title)
bigramas_description_df = pd.DataFrame(bigramas_description)
```

```python

# Definimos los modelos que vamos a utilizar
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Random Forest Regression': RandomForestRegressor()
}

# Ajustamos los hiperparámetros para Ridge y Lasso
param_grid = {
    'Ridge Regression': {'alpha': [0.1, 1, 10]},
    'Lasso Regression': {'alpha': [0.1, 1, 10]}
}

# Realizamos una validación cruzada en 10 partes para cada modelo
for name, model in models.items():
    if name in param_grid:
        grid_search = GridSearchCV(model, param_grid[name], cv=10, scoring='neg_mean_squared_error')
        grid_search.fit(X_train, y_train)
        best_model = grid_search.best_estimator_
    else:
        best_model = model

    # Calculamos el MSE promedio en validación cruzada
    scores = cross_val_score(best_model, X_train, y_train, cv=10, scoring='neg_mean_squared_error')
    rmse_scores = np.sqrt(-scores)
    print(f'Modelo: {name}')
    print(f'MSE Promedio: {np.mean(rmse_scores)}')

# Seleccionamos el modelo con el menor MSE promedio
best_model_name = min(models, key=lambda x: np.mean(np.sqrt(-cross_val_score(models[x], X_train, y_train, cv=10, scoring='neg_mean_squared_error'))))
best_model = models[best_model_name]
print(f'Mejor modelo: {best_model_name}')

# Entrenamos el mejor modelo con todos los datos de entrenamiento
best_model.fit(X_train, y_train)

# Calculamos el MSE en el conjunto de prueba
y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'MSE en conjunto de prueba: {mse}')
print(f'RMSE en conjunto de prueba: {rmse}')

# Obtenemos la importancia de las características (solo para Random Forest)
if isinstance(best_model, RandomForestRegressor):
    feature_importances = best_model.feature_importances_
    print('Importancia de las características:')
    for i, importance in enumerate(feature_importances):
        print(f'Feature {i}: {importance}')
```
